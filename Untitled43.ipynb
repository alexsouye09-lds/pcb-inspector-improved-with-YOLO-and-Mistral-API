````python
from ultralytics import YOLO
import cv2
import json

# 1. Charger ton modèle (prédéfini ou celui que tu as entraîné)
model = YOLO("/besto.pt")  # ton modèle .pt

# 2. Image d'entrée à prédire
image_path = "/kaggle/input/pcb-defects/PCB_DATASET/images/Short/01_short_01.jpg"

# 3. Exécuter la prédiction
results = model(image_path)  # lance l’inférence

# 4. Extraire et structurer les prédictions
predictions = []
for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        conf = float(box.conf[0])
        cls_id = int(box.cls[0])
        # Nom de la classe si disponible
        cls_name = result.names[cls_id] if hasattr(result, "names") else str(cls_id)

        predictions.append({
            "class_id": cls_id,
            "class_name": cls_name,
            "confidence": conf,
            "bbox": [x1, y1, x2, y2]
        })

# 5. Afficher ou enregistrer en JSON
predictions_json = json.dumps({"detections": predictions}, indent=2)
print(predictions_json)

# 6. Optionnel : visualiser l’image avec les boxes YOLO
annotated_image = results[0].plot()  # ajoute des bounding boxes
cv2.imshow("YOLO Preds", annotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
import requests
import json
from ultralytics import YOLO

# === PARTIE 1 : YOLO PREDICTION ===

# Charger ton modèle
model = YOLO("/pcb (4) (4).pt")

# Image à tester
image_path = "/content/pcb-defects/PCB_DATASET/images/Missing_hole/01_missing_hole_06.jpg"

# Faire la prédiction
results = model(image_path)

# Construire le JSON de détections
detections = []
for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        conf = float(box.conf[0])
        cls_id = int(box.cls[0])
        cls_name = result.names[cls_id] if hasattr(result, "names") else str(cls_id)
        detections.append({
            "class_id": cls_id,
            "class_name": cls_name,
            "confidence": conf,
            "bbox": [x1, y1, x2, y2]
        })

pcb_json = {"detections": detections}
print("YOLO JSON:\n", json.dumps(pcb_json, indent=2))

# === PARTIE 2 : APPEL API IA ===

# Construire le prompt pour l'agent
prompt = f"""
Tu es Mistrouspector Chat, un agent expert en inspection PCB.
Analyse ces détections:
{json.dumps(pcb_json, indent=2)}

Répond uniquement en JSON avec les champs:
- instructions (liste de strings)
- priorité (faible / moyenne / élevée)
- analyse_visuelle (description des zones critiques détectées)
"""

# Configuration de l'appel API (format OpenAI compatible)
api_url = "https://api.openai.com/v1/chat/completions"  # ou l'URL Mistral compatible
headers = {
    "Authorization": "Bearer TA_CLE_API",
    "Content-Type": "application/json"
}

data = {
    "model": "mistral-small-latest",  # ou un modèle IA de ton choix
    "messages": [
        {"role": "system", "content": "You are an expert PCB inspection assistant."},
        {"role": "user", "content": prompt}
    ],
    "temperature": 0.3
}

# Faire la requête
response = requests.post(api_url, headers=headers, json=data)
llm_output = response.json()

# Récupérer le texte généré
agent_json_str = llm_output["choices"][0]["message"]["content"]
print("\nAgent IA raw output:\n", agent_json_str)

# Tenter de parser en JSON Python
try:
    agent_output = json.loads(agent_json_str)
    print("\nAgent IA parsed JSON:\n", json.dumps(agent_output, indent=2))
except:
    print("Impossible de parser directement en JSON. Voici le texte brut:")
    print(agent_json_str)
````

````python
from ultralytics import YOLO
from mistralai import Mistral
import json
import os

# -------------------------
# 1) Charger modèle YOLO
# -------------------------
model = YOLO("/pcb (4) (4).pt")

image_path = "/content/pcb-defects/PCB_DATASET/images/Short/01_short_02.jpg"

results = model(image_path)

# -------------------------
# 2) Convertir prédictions en JSON
# -------------------------
detections = []

for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        conf = float(box.conf[0])
        cls_id = int(box.cls[0])
        cls_name = result.names[cls_id] if hasattr(result, "names") else str(cls_id)

        detections.append({
            "class_id": cls_id,
            "class_name": cls_name,
            "confidence": conf,
            "bbox": [x1, y1, x2, y2]
        })

pcb_json = {"detections": detections}

print("=== YOLO OUTPUT ===")
print(json.dumps(pcb_json, indent=2))

# -------------------------
# 3) Appel Agent Mistral
# -------------------------

client = Mistral(api_key=os.environ["MISTRAL_API_KEY"])

system_prompt = """
Tu es un agent AI expert en inspection PCB industrielle.

Tu reçois une liste JSON de défauts détectés.

Ta mission :
- Générer un plan d'action clair
- Prioriser le défaut
- Fournir une analyse technique en étant précis

Répond STRICTEMENT en JSON avec :

{
  "instructions": [liste de strings],
  "priorité": "faible | moyenne | élevée",
  "analyse_visuelle": {
      "défaut": "...",
      "description": "...",
      "localisation": "..."
  }
}
"""

response = client.chat.complete(
    model="mistral-small-latest",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": json.dumps(pcb_json)}
    ],
    temperature=0.2
)

agent_output = response.choices[0].message.content

print("\n=== AGENT OUTPUT RAW ===")
print(agent_output)

# -------------------------
# 4) Parser JSON proprement
# -------------------------
try:
    parsed = json.loads(agent_output)
    print("\n=== AGENT OUTPUT PARSED ===")
    print(json.dumps(parsed, indent=2))
except:
    print("\n⚠️ Réponse non JSON valide.")
````

````python
from ultralytics import YOLO
from mistralai import Mistral
import json
import os
import re

# -------------------------
# 1) Charger modèle YOLO
# -------------------------
model = YOLO("/pcb (4) (4).pt")
image_path = "/content/pcb-defects/PCB_DATASET/images/Short/01_short_02.jpg"

results = model(image_path)

# -------------------------
# 2) Convertir prédictions en JSON
# -------------------------
detections = []

for result in results:
    for box in result.boxes:
        x1, y1, x2, y2 = box.xyxy[0].tolist()
        conf = float(box.conf[0])
        cls_id = int(box.cls[0])
        cls_name = result.names[cls_id] if hasattr(result, "names") else str(cls_id)

        detections.append({
            "class_id": cls_id,
            "class_name": cls_name,
            "confidence": conf,
            "bbox": [x1, y1, x2, y2]
        })

pcb_json = {"detections": detections}

print("=== YOLO OUTPUT ===")
print(json.dumps(pcb_json, indent=2))

# -------------------------
# 3) Appel Agent Mistral
# -------------------------
client = Mistral(api_key=os.environ["MISTRAL_API_KEY"])

system_prompt = """
Tu es un agent AI expert en inspection PCB industrielle.

Tu reçois une liste JSON de défauts détectés.

Ta mission :
- Générer un plan d'action clair
- Prioriser le défaut
- Fournir une analyse technique précise

Répond STRICTEMENT en JSON avec :

{
  "instructions": [liste de strings],
  "priorité": "faible | moyenne | élevée",
  "analyse_visuelle": {
      "défaut": "...",
      "description": "...",
      "localisation": "..."
  }
}
"""

response = client.chat.complete(
    model="mistral-small-latest",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": json.dumps(pcb_json)}
    ],
    temperature=0.2
)

agent_output = response.choices[0].message.content

# -------------------------
# 4) Nettoyer et parser JSON
# -------------------------
# Supprime les backticks ```json ou ``` éventuels
clean_output = re.sub(r"```json|```", "", agent_output).strip()

try:
    parsed = json.loads(clean_output)
    print("\n=== AGENT OUTPUT PARSED ===")
    print(json.dumps(parsed, indent=2))
except Exception as e:
    print("\n⚠️ Erreur parsing JSON :", e)
    print("\nRAW OUTPUT:")
    print(agent_output)
````
